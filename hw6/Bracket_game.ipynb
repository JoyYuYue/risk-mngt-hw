### Statistics about the last four team from 1939 to 2018

```python
raw_four = pd.read_excel('all_four.xlsx',sheetname = 'Sheet1',header=0)
all_four = raw_four[['Year','Score','Winner','Runner-Up','Third Place','Fourth Place']].copy()
all_four.columns = ['Year','Score','First','Second','Third','Fourth']
all_four
```

```python
states = {}
def gettimevalue(n):
    return (n-1938)**2/3
for i in range(len(all_four)):
    first = all_four['First'][i]
    second = all_four['Second'][i]
    third = all_four['Third'][i]
    fourth = all_four['Fourth'][i]
    time_value = gettimevalue(all_four['Year'][i])
    try:
        states[first] += int(4*time_value)
    except:
        states[first] = int(4*time_value)
    try:
        states[second] += int(3*time_value)
    except:
        states[second] = int(3*time_value)
    try:
        states[third] += int(2*time_value)
    except:
        states[third] = int(2*time_value)
    try:
        states[fourth] += int(1*time_value)
    except:
        states[fourth] = int(1*time_value)
import operator
sorted_x = sorted(states.items(), key=operator.itemgetter(1))
sorted_x[::-1]
```

```python
raw_data = pd.read_excel('1819.xlsx',header=1)
a=raw_data.columns
x = list(a)[2:]
data = raw_data[x].copy()
data
```

```python
mean_v=np.mean(data,axis=1)
demeaned_m = data.T - mean_v
pd.DataFrame(demeaned_m)
```

```python
cov_m = pd.DataFrame.dot(demeaned_m.T,demeaned_m) / len(demeaned_m)
pd.DataFrame(cov_m)
```

```python
inv_c = np.linalg.inv(cov_m)
pd.DataFrame(inv_c)
```
