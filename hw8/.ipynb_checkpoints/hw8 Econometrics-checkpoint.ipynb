### Question 1
Transform the series as follows. Think about why these are (or perhaps are not) reasonable transformations, and what assumptions about the data might justify them.

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data=pd.read_csv('NYU Spring 2019 Assignment 8 data.csv',header=0)
data['10-year rate'] = data['10-year rate'].str.rstrip('%').astype('float') / 100.0
data
```

1) Take the natural logarithm of the S&P500 price


Take logarithm transfomation make sense for growth and compounded value.

```python
sp500 = data['S&P500']
sp500_2=np.zeros(len(sp500))
for i in range(len(sp500)):
    sp500_2[i] = math.log(sp500[i])
```

2) Compute the dividend yield as dividend divided by price


```python
dividends = data['Dividends']
df2['Price']=sp500
dividend_yield = np.zeros(len(dividends))
for i in range(len(dividends)):
    dividend_yield[i]=dividends[i]/sp500[i]
```

3) Take the natural logarithm of earnings


```python
earning = data['Earnings']
earning_2=np.zeros(len(earning))
for i in range(len(earning)):
    earning_2[i] = math.log(earning[i])
```

4) Take the natural logarithm of CPI

```python
cpi = data['CPI']
cpi_2=np.zeros(len(cpi))
for i in range(len(cpi)):
    cpi_2[i] = math.log(cpi[i])
```

5) Leave the 10-year rate untransformed

6) Take the natural logarithm of the price of gold

```python
gold = data['Gold']
gold_2=np.zeros(len(gold))
for i in range(len(gold)):
    gold_2[i] = math.log(gold[i])
```

7) Accumulate the President's age, that is each month take the previous month's value (start at zero) and add the current month's value

Don't know why we do this transformation

8) Leave the temperature untransformed


```python
age = data['President age']
age_2 = np.zeros(len(age))
age_2[0]=age[0]
for i in range(1,len(age)):
    age_2[i] = age[i]+age[i-1]
```

```python
result = np.matrix([sp500_2,dividend_yield,earning_2,cpi_2,data['10-year rate'],gold_2,age_2,data['Temperature']]).T
col = data.columns.values[1:]
data_2 = pd.DataFrame(result,columns =col)
data_2
```
